```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('winequality-red.csv', sep=';')

# Fill null values with the mean of their respective columns
df.fillna(df.mean(), inplace=True)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

```python
# Extract features as 1D vectors
alcohol_vector = df['alcohol'].to_numpy()
citric_acid_vector = df['citric acid'].to_numpy()

print("Alcohol Vector Shape:", alcohol_vector.shape)
print("Citric Acid Vector Shape:", citric_acid_vector.shape)
```

```python
# Select features and build the feature matrix X
selected_features = df[['alcohol', 'density']]
X = selected_features.to_numpy()

# Compute the covariance matrix
covariance_matrix = np.cov(X.T)
print("Covariance Matrix:\n", covariance_matrix)
```

```python
# Perform eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
top_eigenvalues = eigenvalues[sorted_indices]
corresponding_eigenvectors = eigenvectors[:, sorted_indices]

print("Top 2 Eigenvalues:", top_eigenvalues)
print("Corresponding Eigenvectors:\n", corresponding_eigenvectors)
```

```python
# Find the most common wine quality score (Mode)
most_common_quality = df['quality'].mode()[0]
print(f"The most common wine quality is: {most_common_quality}")

# Get the proportional distribution (probability distribution)
quality_distribution = df['quality'].value_counts(normalize=True).sort_index()
print("\nDistribution of Wine Quality Scores (Proportions):")
print(quality_distribution)
```
